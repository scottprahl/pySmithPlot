The `pysmithchart` module extends matplotlib by registering a new projection called `smith`. This projection is specifically designed for plotting Smith charts with customized gridlines, axes, and symbols.

### Registration of the "smith" Projection

Simply importing the module will automatically register the `smith` projection with matplotlib. For example:

```python
import pysmithchart
```

After importing, you can use the `smith` projection with any matplotlib axes that support projections.

### Setting the Smith Chart Projection

There are several ways to set up a subplot with the `smith` projection.

- **Using `plt.subplot()`**

````python
import matplotlib.pyplot as plt
import pysmithchart

plt.figure(figsize=(6, 6))
plt.subplot(1, 1, 1, projection="smith")
plt.show()
````

- **Using `plt.subplots()` with the `subplot_kw` parameter:**

````python
  import matplotlib.pyplot as plt
  import pysmithchart

  fig, ax = plt.subplots(1, 1, figsize=(6, 6), subplot_kw={"projection": "smith"})
  plt.show()
````

- **Creating an axes directly via the figures `add_subplot` method:**

````python
import matplotlib.pyplot as plt
import pysmithchart

fig = plt.figure(figsize=(6, 6))
ax = fig.add_subplot(111, projection="smith")
plt.show()
````

### Modifying the Gridlines and Other Parameters

Matplotlib generates the gridlines and other elements of the Smith chart early in the plotting process. This means that any modifications to the grid or other aspects of the Smith chart **must** be specified at the time the subplot (or axes) is created.

The default reference impedance Z = 50.  If this needs to be changed, then this must also be done at the time the subplot is created using `axes_impedance`.

For multiple changes, the `plt.subplot()` command can become unwieldy.  Using a dictionary is a convenient way to mange multiple changes to the Smith chart.

````python
import matplotlib.pyplot as plt
import pysmithchart

plt.figure(figsize=(6, 6))
params = {
    "grid_major_color_x": "blue",
    "grid_major_color_y": "red",
    "grid_minor_enable": True,
    "grid_minor_color_x": "purple",
    "grid_minor_color_y": "orange",
    "axes_impedance": 200,
}
plt.subplot(1, 1, 1, projection="smith", **params)
plt.show()
````
